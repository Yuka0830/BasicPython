# 【確認課題】Matplotlib

```python
# 使うライブラリ
'''
python3.12以降を使用してる場合、japanize-matplotlibは使えません。
その場合、matplotlib-fontjaを使用してください。
'''

import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import matplotlib_fontja
```

## 1問目
使うcsvはhist_data.csvです。下にコードを書いてください。

![問1](./imgs/matplotlib/plot_1.png)

```python
df=pd.read_csv('data/hist_data.csv')
plt.hist(df['aデータ'],bins=30,color='red',alpha=0.5,label="aデータ")
plt.hist(df['bデータ'],bins=30,color='turquoise',alpha=0.5,label="bデータ")
plt.title('aデータとbデータの温度に関する度数分布')
plt.xlabel('温度(℃)')
plt.ylabel('度数')
plt.legend()
```

## 2問目
使うcsvはscatter_data.csvです。下にコードを書いてください。

![問2](./imgs/matplotlib/plot_2.png)

```python
df2=pd.read_csv('data/scatter_data.csv')
plt.scatter(df2[df2['class']==0]['length (cm)'],df2[df2['class']==0]['width (cm)'],label="class0",color='red')
plt.scatter(df2[df2['class']==1]['length (cm)'],df2[df2['class']==1]['width (cm)'],label="class1",color='blue')
plt.scatter(df2[df2['class']==2]['length (cm)'],df2[df2['class']==2]['width (cm)'],label="class2",color='green')
plt.title('クラス0～2のlengthとwitdthに関する散布図')
plt.xlabel('length (cm)')
plt.ylabel('width (cm)')
plt.legend()
plt.grid(True)
```

## 3問目
使うcsvはline_data.csvです。下に2つの画像を表示するコードを書いてください。

![問3_1](./imgs/matplotlib/plot_3_1.png)

![問3_2](./imgs/matplotlib/plot_3_2.png)

```python
df3=pd.read_csv('data/line_data.csv')
x=df3['期間']
y=df3['件数']
plt.plot(x,y,marker="o")
plt.title('期間ごとの件数の推移')
plt.xlabel('期間')
plt.xticks(rotation=30)
plt.ylabel('件数')
plt.grid(True)
```

```python
df4=pd.read_csv('data/line_data.csv')
x=df4['期間']
y=df4['費用']
plt.bar(x,y)
plt.title('期間ごとの費用の推移')
plt.xlabel('期間')
plt.xticks(rotation=30)
plt.ylabel('費用(円)')
plt.grid(True)
```
